```python
# Handle missing values in DataFrame

data = {
    'ID': [1, 2, 3, 1, 2, None, 3, 2, 3, None],
    'Name': ['John', 'Jane', 'Doe', 'John', 'Jane', None, 'Doe', 'Jane', 'Doe', None],
    'Age': [25, 30, 22, 25, 30, 30, 22, None, 22, 25],
    'Salary': [40000, 60000, 80000, 40000, 60000, None, 80000, 60000, 80000, None]
}
# Create DataFrame
df = pd.DataFrame(data)
print(df)
# Step 1: Identify missing values
# print("Missing values per column:\n", df.isnull().sum())

# Option 1: Drop rows with missing values
# df_cleaned = df.dropna()
# df_cleaned

# Option 2: Fill missing values (e.g., forward fill method)
# df_filled = df.ffill()  # Fills missing values with the previous value
# df_filled

# Option 3: Fill missing  values with 0 in place.
df.fillna(0, inplace=True)
# df

# # Convert columns to integers (except for 'Name')
df[['ID', 'Age', 'Salary']]=df[['ID', 'Age', 'Salary']].astype(int)
df
```
